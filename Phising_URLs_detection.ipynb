## chuẩn hóa dữ liệu

```python
# chia tập dữ liệu train test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df['url'], df['detection phishing'], test_size=0.4, random_state=42)
```

```python
# dùng CountVectorizer fit_transform dữ liệu 
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X_CountVectorizer_train = vectorizer.fit_transform(X_train)
X_CountVectorizer_test = vectorizer.transform(X_test)
X_CountVectorizer_train.shape
```

```python
y_train.shape
```

```python
y_train.value_counts()
```

```python
# dùng TfidfVectorizer fit_transform dữ liệu
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer1 = TfidfVectorizer()
X_TfidfVectorizer_train = vectorizer.fit_transform(X_train)
X_TfidfVectorizer_test = vectorizer.transform(X_test)
```

## dùng gridseachCV để tìm siu tham số thích hợp nhất cho multinomialNB


```python
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import MultinomialNB
param_grid = {'alpha': [0.01, 0.1, 1.0, 10.0],
              'fit_prior': [True, False]}

# Define the MultinomialNB classifier
mnb = MultinomialNB()

# Perform grid search cross-validation
grid_search = GridSearchCV(mnb, param_grid, cv=5)
grid_search.fit(X_CountVectorizer_train, y_train)
print("Best parameters: ", grid_search.best_params_)
print("Best score: ", grid_search.best_score_)
# Perform grid search cross-validation
grid_search1 = GridSearchCV(mnb, param_grid, cv=5)
grid_search1.fit(X_TfidfVectorizer_train, y_train)
print("Best parameters: ", grid_search1.best_params_)
print("Best score: ", grid_search1.best_score_)
```

```python
# dùng xây dựng mô hình với MultinomialNB huấn luyện với tập dữ liệu của Countvectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
mnb1 = MultinomialNB(alpha=0.01,fit_prior=True)
mnb1.fit(X_CountVectorizer_train, y_train)

# Dự đoán trên tập kiểm tra và đánh giá hiệu suất
y_pred11 = mnb1.predict(X_CountVectorizer_test)
print('Accuracy:', accuracy_score(y_test, y_pred11))
print('Precision:', precision_score(y_test, y_pred11))
print('Recall:', recall_score(y_test, y_pred11))
print('F1 Score:', f1_score(y_test, y_pred11))
```

```python
# dùng xây dựng mô hình với MultinomialNB huấn luyện với tập dữ liệu của TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
mnb2 = MultinomialNB()
mnb2.fit(X_TfidfVectorizer_train, y_train)
# Dự đoán trên tập kiểm tra và đánh giá hiệu suất
y_pred12 = mnb2.predict(X_TfidfVectorizer_test)
print('Accuracy:', accuracy_score(y_test, y_pred12))
print('Precision:', precision_score(y_test, y_pred12))
print('Recall:', recall_score(y_test, y_pred12))
print('F1 Score:', f1_score(y_test, y_pred12))
```

```python
# boosting hai mô hình mnb2 và mnb1
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

# Tạo một bộ phân loại AdaBoost với DecisionTreeClassifier làm bộ phân loại cơ bản
ada_boost = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=1))

# Huấn luyện mô hình trên tập huấn luyện
ada_boost.fit(X_CountVectorizer_train, y_train)

# Sử dụng mô hình để dự đoán trên tập kiểm tra và tính toán độ chính xác
y_pred = ada_boost.predict(X_CountVectorizer_test)
accuracy = ada_boost.score(X_CountVectorizer_test, y_test)
print("Accuracy:", accuracy)

# y_pred30 = boosted_model.predict(X_CountVectorizer_test)
# print('Accuracy:', accuracy_score(y_test, y_pred12))
# print('Precision:', precision_score(y_test, y_pred12))
# print('Recall:', recall_score(y_test, y_pred12))
# print('F1 Score:', f1_score(y_test, y_pred12))
```

## cũng tương tự cách làm trên nhưng mà LOGISTICS REGRESSION

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

param_grid = {'C': [0.001, 0.01, 0.1, 1, 10],
              'penalty': ['l1', 'l2'],
              'max_iter': [100, 500, 1000],
              'class_weight': [None, 'balanced']}

grid_search = GridSearchCV(LogisticRegression(solver='lbfgs'),
                           param_grid,
                           cv=5,
                           scoring='accuracy',
                           n_jobs=-1)

grid_search.fit(X_CountVectorizer_train, y_train)

print(f"Best parameters: {grid_search.best_params_}")
print(f"Best cross-validation score: {grid_search.best_score_}")
```

```python
# dùng xây dựng mô hình với LogisticRegression huấn luyện với tập dữ liệu của CountVectorizer

logreg = LogisticRegression(C= 0.1, class_weight= None, max_iter= 1000, penalty= 'l2')
lr1 = logreg.fit(X_CountVectorizer_train, y_train)

# Dự đoán trên tập kiểm tra và đánh giá hiệu suất
y_pred21 = lr1.predict(X_CountVectorizer_test)
print('Accuracy:', accuracy_score(y_test, y_pred21))
print('Precision:', precision_score(y_test, y_pred21))
print('Recall:', recall_score(y_test, y_pred21))
print('F1 Score:', f1_score(y_test, y_pred21))

# dùng xây dựng mô hình với LogisticRegression huấn luyện với tập dữ liệu của TfidfVectorizer

lr2 = logreg.fit(X_TfidfVectorizer_train, y_train)

# Dự đoán trên tập kiểm tra và đánh giá hiệu suất
y_pred22 = lr2.predict(X_TfidfVectorizer_test)
print('Accuracy:', accuracy_score(y_test, y_pred22))
print('Precision:', precision_score(y_test, y_pred22))
print('Recall:', recall_score(y_test, y_pred22))
print('F1 Score:', f1_score(y_test, y_pred22))
```
